# LogisticRegression

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed()
from sklearn.linear_model import LogisticRegression
```

```python
x=np.concatenate([np.random.normal(2,1,50),np.random.normal(6,1,50)]).reshape(-1,1)
y=np.array([0]*50+[1]*50)
model=LogisticRegression()
model.fit(x,y)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color="red",label="DataPoints")
x_test=np.linspace (x.min(),x.max(),1000).reshape(-1,1)
y_prob=model.predict_proba(x_test)[:,1]
plt.plot(x_test,y_prob,color="blue",linewidth=2,label="LogisticRegressionCruve")
plt.xlabel("feature")
plt.ylabel("probability")
plt.title('logisticregression')
plt.legend()
plt.show()
```

# LinearRegression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression,LinearRegression
```

```python
#generating a synthetic data
np.random.seed()
x=np.concatenate([np.random.normal(2,1,50),np.random.normal(6,1,50)]).reshape(-1,1)
y=np.array([0]*50+[1]*50)
```

```python
#train a logistic regression model
logistic_model=LogisticRegression()
logistic_model.fit(x,y)
```

```python
#train a linear regression model
linear_model=LinearRegression()
linear_model.fit(x,y)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color="red",label="Datapoints")
x_test=np.linspace(x.min(),x.max(),1000).reshape(-1,1)
y_logistic_prob=logistic_model.predict_proba(x_test)[:,1]
plt.plot(x_test,y_logistic_prob,color="blue",linewidth=2,label="LogisticRegressionCurve")
y_linear_pred=linear_model.predict(x_test)
plt.plot(x_test,y_linear_pred,color="green",linestyle="--",linewidth=2,label="LinearRegressionCurve")
plt.xlabel("feature")
plt.ylabel("output")
plt.title("LOGISTIC VS LINEAR REGRESSION")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv("/content/insurance_data.csv")
data.head()
```

```python
x=data[["age"]]
y=data[["bought_insurance"]]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3)
logistic_model=LogisticRegression()
logistic_model.fit(xtrain,ytrain)
y_pred=logistic_model.predict(xtest)
y_pred
```

```python
logistic_model.score(x,y)
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_pred,ytest)
```

```python
import seaborn as sns
sns.heatmap (c,annot=True)
plt.xlabel("predict")
plt.ylabel("True")
plt.figure(figsize=(20,7))
```
